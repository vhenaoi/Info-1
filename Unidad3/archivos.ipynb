```python
try:
    # Paso 1: Abrir un archivo en modo lectura ("r")
    arch = open("archivo1.txt","r")

    # Paso 2: leer el archivo
    for linea in arch:
        #es el último carater un salto?
        if linea[-1] == '\n':
            #entonces solo muestro del prinicipio
            #hasta el caraterter que este antes del ultimo.
            linea = linea[:-1]
        print(linea)

    # Paso 3: cerrar el archivo

    arch.close()

except IOError:
    print("Ha ocurrido un error al intentar abrir el archivo ")
```

```python
try:
    # Paso 1: Abrir un archivo en modo lectura ("r")
    arch = open("archivo1.txt","r")

    caracter =  arch.read(1)

    contador = 0

    while caracter != '':

        contador += 1
        caracter = arch.read(1)

    arch.close()

    print("Cantidad de carateres del archivo: %d"%contador)

except IOError:
    print("Ha ocurrido un error al intentar abrir el archivo ")
```

```python
try:
    arch = open("archivo1.txt", "r")
    linea = arch.readline()

    while linea != '':
        if linea[-1] == "\n":
            linea = linea[:-1]
        print(linea)
        linea = arch.readline()

    arch.close()
except IOError:
    print("Ha ocurrido un error al intentar abrir el archivo ")
```

```python
try:
    arch = open("multiplicar.txt","w")
    num = int(input("Ingrese un número: "))

    for i in range(11):
        arch.write(num*i)

    arch.close()

except IOError:
    print("Erro al escribir el archivo")
```
